```r
# plot_upsetr_dr_and_sav.ipynb
"""""""""""""""""""""""""""""""""
# Author: Keila Velazquez-Arcelay
# 
# Description: Plots two UpsetR graphs in R: 
#              1) Number of circadian genes that are divergently regulated (PrediXcan) 
#              in a GTEx tissue for each of the 3 archaic hominins, 
#              2) Number of circadian splice-altering variants that are fixed in each 
#.             of the 4 archaic hominins. 
#
"""""""""""""""""""""""""""""""""
# 
# Setup R kernel: https://izoda.github.io/site/anaconda/r-jupyter-notebook/
# > R
# > install.packages('UpSetR')
# >
# UpSetR: https://cran.r-project.org/web/packages/UpSetR/vignettes/basic.usage.html
# UpSetR: https://cran.r-project.org/web/packages/UpSetR/vignettes/set.metadata.plots.html
# UpSetR: https://www.r-bloggers.com/hacking-our-way-through-upsetr/
# 
```

```r
library('UpSetR')

archaic_dr <- read.csv('../data/circadian_genes_predixcan_dr.tsv', header=TRUE, sep = '\t')
archaic_sav <- read.csv('../data/circadian_variants_sav.tsv', header=TRUE, sep = '\t')

```

```r
# PLOT DIVERGENTLY REGULATED GENE (PREDIXCAN) X GTEX TISSUE PAIRS BY ARCHAIC INDIVIDUAL

#pdf('plots/upset_dr.pdf')
upset(archaic_dr, 
      sets = c('Denisova', 'Vindija', 'Altai'), keep.order = TRUE, 
      order.by=c('freq'), matrix.color='black', point.size=5, 
      mainbar.y.label = 'DR GENE/TISSUE PAIRS',# sets.x.label = "Number of associations",
      text.scale = c(2.5, 2, 2, 2, 2, 2))

#      sets.bar.color=c("maroon","blue","orange","green"))
#      colors = c(gwas = "green", phewas = "navy", gtex = "purple"))
dev.off()
```

```r
# PLOT ARCHAIC-SPECIFIC SPLICE-ALTERING VARIANTS BY ARCHAIC INDIVIDUAL

#pdf('plots/upset_sav.pdf')
upset(archaic_sav, 
      sets = c('Denisovan', 'Chagyrskaya', 'Vindija', 'Altai'), keep.order = TRUE, 
      order.by=c("freq"), matrix.color="black", point.size=5, 
      mainbar.y.label = 'SPLICE ALTERING VARIANTS',# sets.x.label = "Number of associations",
      text.scale = c(2.5, 2, 2, 2, 2, 2))

#      sets.bar.color=c("maroon","blue","orange","green"))
#      colors = c(gwas = "green", phewas = "navy", gtex = "purple"))
dev.off()
```

```python
# plot_morningness_direction_of_effect.ipynb
"""""""""""""""""""""""""""""""""
# Author: Keila Velazquez-Arcelay
# 
# Description: Plot the cumulative fraction of morningness propensity from a GWAS study
#              on the UKBiobank data (Nealelab).
#
"""""""""""""""""""""""""""""""""


# INPUT FILES
filenames = ['../data/plotting_morningness_cumulative_fraction_browning.tsv',
             '../data/plotting_morningness_cumulative_fraction_browningc.tsv',
             '../data/plotting_morningness_cumulative_fraction_i2.tsv', 
             '../data/plotting_morningness_cumulative_fraction_i2c.tsv',
             '../data/plotting_morningness_cumulative_fraction_iall.tsv',
             '../data/plotting_morningness_cumulative_fraction_iallc.tsv']


import pandas as pd
import math
import matplotlib.pylab as plt
import seaborn as sns
sns.set(rc = {'figure.figsize':(15,8)})
sns.set_style('whitegrid')
sns.set_context('poster')
```

```python
def plot_cumulative_fraction(df1, df2, filename):
    plt.axhline(y=0.5, color='silver')
    plt.axvline(x=math.log10(0.00000005), color='red', linewidth=1, linestyle='--', 
                label="Genome-wide significance threshold")
    plt.plot(df1['P-Value(log10)'], df1['Cumulative_Fraction'], linewidth=1.5, marker='o', markersize=5,
            color='dimgray', alpha=0.8, label="Morning person associations")
    plt.plot(df2['P-Value(log10)'], df2['Cumulative_Fraction'], linewidth=1.5, marker='^', markersize=5,
            color='blue', alpha=0.8, label="Morning person associations in circadian genes")
    plt.xlabel('P-Value(log10)')
    plt.ylabel('Morningness_Fraction')
    plt.legend(loc='lower left',fontsize='15')
    plt.xticks(rotation=90)
    #plt.savefig(filename, dpi=500)
    plt.show()
    
```

```python
# LOAD DATA
dataframes = []
for filename in filenames:
    df = pd.read_csv(filename, sep='\t')
    dataframes.append(df)
```

```python
dataframes[0]
```

```python
# PLOT THE CUMULATIVE FRACTION OF MORNINGNESS PROPENSITY ON A SET OF INTROGRESSED SNPS
# IDENTIFIED BY BROWNING ET AL., 2018
plot_cumulative_fraction(dataframes[0], dataframes[1], 
                        '../plots/morningness_introgressed_direction_of_effect_browning+1_ld50.pdf')
```

```python
# PLOT THE CUMULATIVE FRACTION OF MORNINGNESS PROPENSITY ON A SET OF INTROGRESSED SNPS
# IDENTIFIED BY BROWNING ET AL., 2018 AND AT LEAST ONE OTHER METHOD
plot_cumulative_fraction(dataframes[2], dataframes[3], 
                        '../plots/morningness_introgressed_direction_of_effect_browning+1_ld50.pdf')
```

```python
# PLOT THE CUMULATIVE FRACTION OF MORNINGNESS PROPENSITY ON A SET OF INTROGRESSED SNPS
# IDENTIFIED BY BROWNING ET AL., 2018 AND FIVE OTHER METHODS
plot_cumulative_fraction(dataframes[4], dataframes[5], 
                        '../plots/morningness_introgressed_direction_of_effect_browning+5_ld50.pdf')
```
